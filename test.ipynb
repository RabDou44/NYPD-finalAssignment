```python
# testowanie ludosc 
# calculate per population
import pandas as pd
from Pitmodule.population import PopulationFrame
populacje2020 = PopulationFrame()
populacje2020.data["Gminy"]
```

```python
populacje2020.data["Powiaty"] # 314 powiatow oraz 66 M_NPP
powiaty = populacje2020.data["Powiaty"]
m_powiaty = powiaty.loc[powiaty["nazwa"].str.contains("M\.")]
# m_powiaty.loc[m_powiaty['ludnosc'] >70000]
m_powiaty.reset_index(inplace =True)
m_powiaty  = m_powiaty.drop('index',axis= 1)
m_powiaty
```

```python
gminy = populacje2020.data["Gminy"]
gminy['nazwa'] = gminy['nazwa'].astype(str)
m_powiaty['nazwa'] = m_powiaty['nazwa'].astype(str)
pd.merge(left =gminy,right=m_powiaty, on ='ludnosc',how="inner",sort=True)
# inner join na nazwach nie działa (hipoteza: polskie znaki)

test = m_powiaty.loc[[1,4]]
gminy.merge(right=test['nazwa'], on ='nazwa',how="inner",sort=True)
```
